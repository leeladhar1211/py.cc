### One Sample T-Test Hypothesis Testing

```python
ban_mar_age_it_male = [28,29,35,32,25,28,28,28,29,32,31]
```

```python
stats.skew(ban_mar_age_it_male)
```

```python
stats.kurtosis(ban_mar_age_it_male)
```

#### Hypothesis setup
H0 -> H null -> Marriage age IT bangalore Male is 32 <br/>
Ha -> H alt -> Marriage age IT bangalore Male is NOT 32

```python
from scipy import stats
```

```python
stats.ttest_1samp(ban_mar_age_it_male,32)
```

P value = 1.3%, so < 5%, So we reject Hnull and Accept Ha <br/>
Marriage age of IT bangalore Male is NOT 32.

### Two Sample T-Test

```python
ban_mar_age_it_male = [28,29,35,32,25,28,28,28,29,32,31]
che_mar_age_it_male = [27,25,25,32,29,23,25,29,23,30,22]
```

H0 -> Hnull -> Always equals samples <br/>
H0 -> IT Male marriage age in Ban and Che are same <br/>
Ha -> IT Male marriage age in Ban and Che are NOT same <br/>

```python
stats.ttest_ind(ban_mar_age_it_male,che_mar_age_it_male)
```

P value = 2.2%, < 5%, We REJECT H0 Null, <br/>
Means, Ban and Che IT male marriage ages are not SAME.

## Two Sample Relational T-Test

```python
before_red_bull = [36,32,28,23,27,23]
after_red_bull = [18,12,10,12,11,13]
```

##### Setting Hypothesis
H0 -> H null -> Res time before and after are same <br/>
Ha -> H alt -> Res time before and after are NOT same


```python
stats.ttest_rel(before_red_bull,after_red_bull)
```

p=0, which means we REJECT H0, ie., REDBULL has EFFECT

## One Way Anova (Analysis of Variance)

```python
lemon = [36,32,28,23,27,23]
coffee = [23,26,28,20,18,21]
redbull = [18,12,10,12,11,13]
```

### Setting Hypothesis
H0 -> Res time of Lemon, coffee, redbull are same. <br/>
Ha -> Res time of Lemon, coffee, redbull are NOT same.

```python
stats.f_oneway(lemon,coffee,redbull)
```

P = 0, < 5%, We reject NULL hypothesis, <br/>
Which means, Drinks have significant effect of res time
